# Streaming & Batching processing

### Streaming

#### Most model can stream their output content while it is being generated. By displaying output progressively, streaming significantly improves user experience, particularly for longer responses.

#### Call stream() returns an iterator that yields output chunks as they are produced. Iteratively each chunk can be processed real-time.

```python
ai_brief=model.invoke("Write me a 200 words paragraph on Artificial Intelligence")
print(ai_brief.content)
```

```python
for chunk in model.stream("Write me a 200 words paragraph on Artificial Intelligence"):
    print(chunk.text, end="|", flush=True)
```

```python
for chunk in model.stream("Why do parrots have beautiful feathers?"):
    print(chunk.text, end=" ", flush=True)
```

### Batch

#### Batching a collection of independent requests to a model can significantly improve performance and reduce costs as they process parallel.

```python
responses=model.batch([
    "Why do parrots have colored feathers?",
    "How do airplanes fly?",
    "What is quantum computing?"
])

for idx,response in enumerate(responses):
    print(f"Question {idx+1}: {response.content}")
```

```python
responses=model.batch([
    "Why do parrots have colored feathers?",
    "How do airplanes fly?",
    "What is quantum computing?"],
    config={"max_concurrency":5}    
)

for idx,response in enumerate(responses):
    print(f"Question {idx+1}: {response.content}")
```
